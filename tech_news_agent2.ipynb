```python
from dotenv import find_dotenv, load_dotenv
import os
from crewai import Agent, Task, Crew, Process
from crewai_tools import SerperDevTool
from langchain_openai import ChatOpenAI
from langchain_google_genai import ChatGoogleGenerativeAI
from crewai_tools import SerperDevTool
# import streamlit as st

# Replace "YOUR_API_KEY" with your actual Google Cloud Project API Key
# model = ChatGoogleGenerativeAI(model="gemini-pro", api_key="YOUR_API_KEY")
# Optionally, configure model parameters like temperature or top_p
# model = ChatGoogleGenerativeAI(model="gemini-pro", api_key="YOUR_API_KEY", temperature=0.5, top_p=0.9)
SERPER_API_KEY = "f58caf357ea4459bf49ecd99ee34f7a39354ae53"
search_tool = SerperDevTool()

# This agent will be responsible for fetching news articles from chosen sources.
researcher = Agent (
    role = "News Reader",
    goal = "Look up the latest tech related news",
    backstory = """You are an equity sales assistant for the Asia equity tech sales team.
    You are very good at finding tech related news that are relevant to the Asia equity market.""",
    verbose = False,
    allow_delegation=False,
    tools = [search_tool],
    llm = ChatGoogleGenerativeAI(model="gemini-pro", api_key="AIzaSyBMKd3vOEokkqYzvQUqRFk6DM8ho26Ajlg", temperature=0.5, top_p=0.9)
)

writer = Agent (
    role = "News Summerizer",
    goal = "Summerize the tech news findings",
    backstory = """You are an equity sales.
    You are very good at condensing information into easy, concise and digestible content for investors to read.""",
    verbose = False,
    allow_delegation = True,
    llm = ChatGoogleGenerativeAI(model="gemini-pro", api_key="AIzaSyBMKd3vOEokkqYzvQUqRFk6DM8ho26Ajlg")
)

task1 = Task(
    description = """Conduct a Your job is to look through the given list of news websites,
    find any tech related news within the last 12 hours,
    and identify the ones that could potentially have an impact on the Asia equity market.""",
    expected_output = "Full analysis report in bullet points",
    agent = researcher
)

task2 = Task(
    description = """Using the insights provided, develop an engaging blog""",
    expected_output = "Full blog post of at least 4 paragraphs",
    agent = writer
)

# Instantiate your crew with a sequential process
crew = Crew(
    agents = [researcher, writer],
    tasks = [task1, task2],
    verbose = 1, # You can set it to 1 or 2 to different logging levels

)

result = crew.kickoff()

print("############")
print(result)
```
